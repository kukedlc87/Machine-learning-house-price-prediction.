```python
def convertir_valor(valor):
    if isinstance(valor, str) and '>' in valor:
        return int(valor.replace('>', ''))
    else:
        return int(valor)

# Aplicar la función a la columna 'Columna1'
data['Bathroom'] = data['Bathroom'].apply(convertir_valor)
data['Balcony'] = data['Balcony'].apply(convertir_valor)
```

# I took some parts of code from another user, his user names is MITTA SAI NITISH!

```python
# 1. Fill null values in 'Description', 'Facing', 'Overlooking', and 'Ownership' with 'Unknown'
cols_to_fill_unknown = ['Description', 'facing', 'Ownership']
data[cols_to_fill_unknown] = data[cols_to_fill_unknown].fillna('Unknown')

# Convert 'Amount(in rupees)' to numerical format
def convert_amount(amount):
    try:
        if 'Lac' in amount:
            amount = amount.replace('Lac', '').strip()
            return float(amount) * 100000  # Convert Lac to rupees (1 Lac = 100000 rupees)
        elif 'Cr' in amount:
            amount = amount.replace('Cr', '').strip()
            return float(amount) * 10000000  # Convert Cr to rupees (1 Cr = 10000000 rupees)
        else:
            return float(amount)
    except ValueError:
        return None

data['Amount(in rupees)'] = data['Amount(in rupees)'].apply(convert_amount)
```

```python
def convert_carpet_area(area):
    try:
        if isinstance(area, str):
            if 'sqft' in area:
                area = area.replace('sqft', '').strip()
                return float(area)  # Convert to square feet
            elif 'sqm' in area:
                area = area.replace('sqm', '').strip()
                return float(area) * 10.7639  # Convert square meters to square feet (1 sqm = 10.7639 sqft)
        return float(area)  # For numeric values, directly return the converted float
    except ValueError:
        return None

# Convertir toda la columna 'Carpet Area' utilizando la función apply
data['Carpet Area'] = data['Carpet Area'].apply(convert_carpet_area)

# Si lo deseas, puedes llenar los valores nulos con 0 después de la conversión
data['Carpet Area'] = data['Carpet Area'].fillna(0)
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
data.location = le.fit_transform(data.location)
data.Furnishing = le.fit_transform(data.Furnishing)
data.Transaction = le.fit_transform(data.Transaction)
data.Bathroom = data.Bathroom.astype(int)
data.Balcony = data.Balcony.astype(int)
data.Ownership = le.fit_transform(data.Ownership)
data.facing = le.fit_transform(data.facing)




plt.figure(figsize=(12,10))
sns.heatmap(data.corr(), annot=True, cmap='jet')
```

# Predictions!!!

```python
data['Floor'] = data['Floor'].fillna(0)
data['Floor'].isnull().sum()
```

```python
data[['n_floor', 'total_floor']] = data['Floor'].str.split(' out of ', expand=True)

# Replace 'Ground' with 0 in the 'n_floor' and 'total_floor' columns
data['n_floor'] = data['n_floor'].replace('Ground', 0)
data['total_floor'] = data['total_floor'].replace('Ground', 0)
data['n_floor'] = data['n_floor'].replace('Upper Basement', 100)
data['total_floor'] = data['total_floor'].replace('Upper Basement', 100)
data['n_floor'] = data['n_floor'].replace('Lower Basement', 100)
data['total_floor'] = data['total_floor'].replace('Lower Basement', 100)
data['n_floor'] = data['n_floor'].fillna(0)
data['total_floor'] = data['total_floor'].fillna(0)

# Convert the columns to integers
data['n_floor'] = data['n_floor'].astype(int)
data['total_floor'] = data['total_floor'].astype(int)

# Eliminar la columna original
data = data.dropna(subset=['Amount(in rupees)'])
data['Amount(in rupees)'].isnull().sum()
data = data.dropna(subset=['Price (in rupees)'])
data['Price (in rupees)'].isnull().sum()
print(data.info())
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

X = data.drop(['Amount(in rupees)','Floor','Status', 'Description', 'Title'], axis=1)
y = data['Amount(in rupees)']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)

model = XGBRegressor(n_estimators=100)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Calcular las métricas de evaluación
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Imprimir las métricas
print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("R-squared (R2):", r2)
```
